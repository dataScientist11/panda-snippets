### Video 2

```python
print('creating a basic np array')
list1 = [1,2,3, 4, 5]

nplist1 = np.array(list1, dtype=np.float64)
nplist2 = np.array([nplist1 for i in range(3)], dtype=np.float64)
nplist3 = np.array([nplist2 for i in range(4)], dtype=np.float64)
print('nplist1', nplist1)
print('nplist2', nplist2)
print('nplist3', nplist3)
print('\n\nFunctionalities')
print('number of dimensions', nplist1.ndim)
print('number of dimensions', nplist2.ndim)
print('number of dimensions', nplist3.ndim)
print('shape', nplist3.shape)
print('numpy size', nplist3.size)
print('numpy itemsize', nplist3.itemsize)
print('numpy memory taken', nplist3.size*nplist3.itemsize)
print('dtype', nplist3.dtype, nplist3.dtype.name)
print('type', type(nplist3))



```

### creating a normal function


```python
def norm_function(x,mean, var):
    return (1.0/(2.0*np.pi*var)**.5)*np.exp(-(x-mean)**2.0/(2*(var)))
x = np.arange(-5, 5.01, 0.05)
pdf1 = norm_function(x, 0, 0.2)
pdf2 = norm_function(x, 0, 1)
pdf3 = norm_function(x, 0, 5)
pdf4 = norm_function(x, -2, .5)

plt.plot(x,pdf1, color='blue', label='mean = {} var={}'.format(0,.2))
plt.plot(x,pdf2, color='red', label='mean = {} var={}'.format(0,1))
plt.plot(x,pdf3, color='yellow', label='mean = {} var={}'.format(0,5))
plt.plot(x,pdf4, color='green', label='mean = {} var={}'.format(-2  ,.5))
```
